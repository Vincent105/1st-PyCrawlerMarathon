## 根據範例，取出今天所有的新聞

```python
from selenium import webdriver
from bs4 import BeautifulSoup
import time

browser = webdriver.Chrome(executable_path='chromedriver\chromedriver.exe')
browser.get("https://www.ettoday.net/news/news-list.htm")

formatted_today = time.strftime("%Y/%m/%d")
# print(formatted_today)
while True:
    time.sleep(2)
    browser.execute_script("window.scrollBy(0,10000)")
    html_source = browser.page_source
    soup = BeautifulSoup(html_source, "html5lib")
    div_find_h3 = soup.find(class_="part_list_2").find_all('h3')
    for i in div_find_h3:
        if str(i.span.text[:10]) != formatted_today:
            break        
for i in div_find_h3:                
    print(i.span.text,i.em.text,i.a.text)
browser.quit();            
```
